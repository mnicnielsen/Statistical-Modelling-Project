Build a regression model.

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
import pandas as pd
```

```python
biz_and_station_info = pd.read_csv("C:/Users/mnicn/local_documents/lighthouse-data-notes/Week 3/Project 2 Fresh Files/data/biz_and_station_info.csv")
biz_and_station_info = biz_and_station_info.drop(labels="station_id",axis=1)
biz_and_station_info
```

Provide model output and an interpretation of the results. 

```python
#Where there were no businesses within 100 m, this table shows NaN for these stations. 
#Remove for the linear regression
no_nans = biz_and_station_info.dropna()
no_nans
```

```python
#Model to predict availability of bikes 
y = no_nans['Available Bikes']
X = no_nans[['review_count', 'rating']]
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
#Create an aggregated table which counts the number of restaurants by each bike station. b
count_of_businesses = biz_and_station_info.groupby("Station ID")["id"].count()
count_of_businesses = pd.DataFrame(count_of_businesses)
count_of_businesses
```

```python
count_of_businesses = count_of_businesses.reset_index()
count_of_businesses.head(3)
```

```python
count_of_businesses.columns = ['Station ID', 'count of businesses']
count_of_businesses.head(3)
```

```python
#import data on bike stations
stations = pd.read_csv("C:/Users/mnicn/local_documents/lighthouse-data-notes/Week 3/Project 2 Fresh Files/data/df_stations.csv")
stations.head(2)
```

```python
table_with_business_counts = count_of_businesses.merge(stations, how='inner', on='Station ID')
table_with_business_counts.head(3)
```

```python
#Model to predict availability of bikes 
y = table_with_business_counts['Proportion of Bikes Available (%)']
X = table_with_business_counts['count of businesses']
X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```
