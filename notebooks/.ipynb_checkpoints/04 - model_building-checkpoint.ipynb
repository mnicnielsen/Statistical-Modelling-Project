Build a regression model.

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
import pandas as pd
```

```python
fs_df = pd.read_csv("fs_venues.csv")
```

```python
fs_df.head(2)
```

```python
y = fs_df["Total - All Venues"]
X = fs_df.drop("Total - All Venues", axis=1)
X = fs_df.drop("Station ID", axis=1)
X = sm.add_constant(X) #adds a column of 1's so the model will contain an intercept

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
fs_df.head()
```

```python
#Further let's predict the number of bars in an area based on number of other business types in an area. 
y = fs_df["Bars - All"]
X = fs_df.drop(["Total - All Venues", "Station ID", "Bars - All"], axis=1)
X = sm.add_constant(X) #adds a column of 1's so the model will contain an intercept

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```
