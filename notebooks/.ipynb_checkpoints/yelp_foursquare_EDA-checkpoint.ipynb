# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 
Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)
Put your parsed results into a DataFrame

Note: these are all done in a single step in the code below. 

```python
#Make fs GET request
fs_url = 'https://api.foursquare.com/v3/places/search'
fs_key = open(r"C:\Users\mnicn\local_documents\lighthouse-data-notes\Week 3\fs.txt").readlines()[0]
fs_headers = {
    "accept": "application/json",
    "Authorization": fs_key
}
#Set timestamp for Friday 2 pm
#fs_timestamp_fri = 1667595600
fs_timestamp_fri = '5T1400'
```

```python
#Create list with all results.
#Each time the OUTSIDE loop runs, it will put data into one row.
    #Note: in plain english, the business types in the list below (in order are):
        #Dining and Drinking > Bar, Dining and Drinking > Restaurant, Retail, Community and Government > Education, Arts and Entertainment
        #See category descriptions here: https://location.foursquare.com/places/docs/categories
fs_biz_types = ['13003','13065','17000','12009','10000']
fs_k = 0
fs_all_vals = []
while fs_k < test_df.shape[0]:
    fs_lat = str(test_df.iloc[fs_k][1])
    fs_long = str(test_df.iloc[fs_k][2])
    ll = fs_lat + ',' + fs_long

    fs_i = 0
    while fs_i < len(fs_biz_types):
    #Do 2 API requests within each round of the while loop, one for all businesses of a certain type, one for all open businesses of a certain type.
    #While loop to run 5 times (length of biz_types list)

        #Set parameters for ALL businesses of certain type
        fs_parameters_all = {
            'll' : ll,
            'radius':100,
            'limit':50,
            'fields': 'name',
            'categories': fs_biz_types[fs_i]
        }
        #Make API request and make JSON readable
        fs_response = requests.get(fs_url, headers=fs_headers, params=fs_parameters_all)
        fs_num_biz_all = len(json.loads(fs_response.text)['results'])
        #test_df.iloc[k][i+3] = num_biz_all
        fs_all_vals.append(fs_num_biz_all)

        #Set parameters for OPEN businesses of certain type
        fs_parameters_open = {
            'll' : ll,
            'radius':100,
            'limit':50,
            'fields': 'name',
            'open_at': fs_timestamp_fri,
            'categories': fs_biz_types[fs_i]
        }
        #Make API request and make JSON readable
        fs_response = requests.get(fs_url, headers=fs_headers, params=fs_parameters_open)
        fs_num_biz_open = len(json.loads(fs_response.text)['results'])
        fs_all_vals.append(fs_num_biz_open)
        fs_i += 1
    fs_k+=1
#print(all_vals)
fs_reshaped_vals = np.reshape(fs_all_vals, (fs_k,10))
fs_df_vals = pd.DataFrame(fs_reshaped_vals)
fs_df = pd.merge(test_df, fs_df_vals, left_index=True, right_index=True)

fs_all_col_names = ['Station ID', 'latitude', 'longitude', 'Dining and Drinking > Bar - All', 'Dining and Drinking > Bar - Open', 'Dining and Drinking > Restaurant - All', 'Dining and Drinking > Restaurant - Open', 'Retail - All', 'Retail - Open', 'Community and Government > Education - All', 'Community and Government > Education - Open', 'Arts and Entertainment - All', 'Arts and Entertainment - Open']
fs_df.columns = fs_all_col_names
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)
Put your parsed results into a DataFrame

```python
#Make Yelp GET request
url = 'https://api.yelp.com/v3/businesses/search'
key = open(r'C:\Users\mnicn\local_documents\lighthouse-data-notes\Week 3\Yelp2.txt').readlines()[0]
headers = {
    'Authorization': 'Bearer %s' % key
}
#Set timestamp for Friday 2 pm
timestamp_fri = 1667595600
```

```python
#Create list with all results.
#Each time the OUTSIDE loop runs, it will put data into one row. 
biz_types = ['nightlife, All','restaurants, All','shopping, All','education, All','arts, All']
k = 0
all_vals = []
while k < test_df.shape[0]:
    lat = test_df.iloc[k][1]
    long = test_df.iloc[k][2]

    i = 0
    while i < len(biz_types):
    #Do 2 API requests within each round of the while loop, one for all businesses of a certain type, one for all open businesses of a certain type.
    #While loop to run 5 times (length of biz_types list)

        #Set parameters for ALL businesses of certain type
        parameters_all = {
            'latitude' : lat,
            'longitude' : long,
            'radius':100,
            'limit':50,
            'categories': biz_types[i]
        }
        #Make API request and make JSON readable
        response = requests.get(url, headers=headers, params=parameters_all)
        num_biz_all = len(json.loads(response.text)['businesses'])
        #test_df.iloc[k][i+3] = num_biz_all
        all_vals.append(num_biz_all)

        #Set parameters for OPEN businesses of certain type
        parameters_open = {
            'latitude' : lat,
            'longitude' : long,
            'radius':100,
            'limit':50,
            'open_at': timestamp_fri,
            'categories': biz_types[i]
        }
        #Make API request and make JSON readable
        response = requests.get(url, headers=headers, params=parameters_open)
        num_biz_open = len(json.loads(response.text)['businesses'])
        all_vals.append(num_biz_open)
        i += 1
    k+=1
#print(all_vals)
reshaped_vals = np.reshape(all_vals, (k,10))
df_vals = pd.DataFrame(reshaped_vals)
yelp_df = pd.merge(test_df, df_vals, left_index=True, right_index=True)

all_col_names = ['Station ID', 'latitude', 'longitude', 'nightlife, All - total','nightlife, All - open','restaurants, All - total','restaurants, All - open', 'shopping, All - total','shopping, All - open', 'education, All - total','education, All - open', 'arts, All - total', 'arts, All - open']
yelp_df.columns = all_col_names
```
